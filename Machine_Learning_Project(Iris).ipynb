```python
s = {'setosa':0,'versicolor':1,'virginica':2}
```

```python
df["species"] = df["species"].map(s)
```

```python
df.sample(5)
```

```python
x = df.drop("species",axis=1)
y = df["species"]
```

```python
x.head()
```

```python
df.isna().sum()
```

```python
# How to divide the data for training and testing
from sklearn.model_selection import train_test_split
```

```python
# 100% =   Train  80%     Test  20%
#            x,y           x,y

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Logistic Regression : It is one of the most popular machine learning algorithm, which comes under the supervised machine learning
#                       It is used to predict the categoical dependent variable
#                       It works on binary classification probem like 0,1
```

```python
# What is sigmoid function in linear regression?
# It is mathematically function used to map the predicted values to probabilities
# We use sigmoid function for binary classification
# This function map any real value between 0 and 1 in machine leaning
```

```python
# Types of Logistic Regression:
#     1. Binary Logistic Regression
#     2. Multinomial Logistic Regression
#     3. Ordinal Logistic Regression
```

```python
# Build a machine learning model which is used for classification of LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
pred = log.predict(x_test)
print(pred)
```

```python
s
```

```python
ac = log.score(x_train,y_train)           # Show the accuracy of training
print(ac)
```

```python
ac2 = log.score(x_test,y_test)             # Show the accuracy of testing data
print(ac2)
```

```python
# How to check the mistake of model in classification?
# You will create a confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
# Confusion matrix shows the performance and mistake of classification model
```
